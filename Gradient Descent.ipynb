```python
# the following file implement the traversal to the minima of a loss function

from IPython.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))

import numpy as np, seaborn as sns, matplotlib.pyplot as plt, time
```

```python
# visualize the dataset
def plot_scatter(x, y, xs=None, ys=None, plot_descent=False):

    plt.figure(figsize=(20, 7))

    sns.set_style('white')
    sns.set_context('paper')

    sns.lineplot(x=x, y=y)

    if plot_descent:
        sns.scatterplot(x=xs, y=ys, c='r')
        plt.title(f'The Gradient Descent', **{'fontname': 'monospace'})
    else:
        plt.title(f'The Dataset', **{'fontname': 'monospace'})

    plt.xticks(**{'fontname': 'monospace'})
    plt.yticks(**{'fontname': 'monospace'})
    plt.show()

    return None

# get derivative at x for a function using the traditional definition of derivatives
get_derivative = lambda x, f, delta: (f(x + delta) - f(x)) / delta

# traverse to the minima of the function
def gradient_descent(x, f, lr, delta):
    
    x_history, grad_history = [], []
    grad = round(get_derivative(x, f, delta), 2)
    
    while abs(grad) > 0.01:
        x_history.append(x)
        grad_history.append(grad)
        
        x -= lr * grad
        grad = get_derivative(x, f, delta)
    
    x_history.append(x)
    grad_history.append(grad)
    
    return x_history, grad_history
```

```python
# define the function to traverse
f = lambda x: x**2
x = np.arange(-10, 10, 0.01)

# plot f(x)
# plot_scatter(x, f(x), False)

# plot gradient descent
x_history, grad_history = gradient_descent(5, f, 0.2, 0.00001)
plot_scatter(x, f(x), x_history, f(np.array(x_history)), True)
```

```python
# define the function to traverse
f = lambda x: np.log(1 + np.exp(np.sin(x)))
x = np.arange(-10, 10, 0.001)

# plot f(x)
# plot_scatter(x, f(x), False)

# plot gradient descent
x_history, grad_history = gradient_descent(6, f, 0.7, 0.1)
plot_scatter(x, f(x), x_history, f(np.array(x_history)), True)
```
